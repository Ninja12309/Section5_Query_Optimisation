# Connect to MongoDB Atlas

```python
from pymongo import MongoClient

connection_string = "mongodb+srv://kunal:kunal@cluster0.l6r5e.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

client = MongoClient(connection_string)

db = client["northstar_mobility"]

orders_col = db["orders"]

complaints_col = db["complaints"]

drivers_col = db["drivers"]

app_events_col = db["app_events"]

print("Connected to database")
```

# Query without index

Example query:

find high priority orders

```python
query = {"priority_level":"High"}

explain_before = orders_col.find(query).explain()

print(explain_before)
```

# Create index

Index created on priority_level field.

```python
orders_col.create_index("priority_level")

print("Index created")
```

# Query after index

```python
explain_after = orders_col.find(query).explain()

print(explain_after)
```

# Index on complaint severity

```python
complaints_col.create_index("severity")

query_complaint = {"severity":"High"}

explain_complaint = complaints_col.find(query_complaint).explain()

print(explain_complaint)
```

# Compound index

```python
complaints_col.create_index("severity")

query_complaint = {"severity":"High"}

explain_complaint = complaints_col.find(query_complaint).explain()

print(explain_complaint)
```

# Query using compound index

```python
query_compound = {

"service_type":"Parcel",

"priority_level":"High"

}

explain_compound = orders_col.find(query_compound).explain()

print(explain_compound)
```

# Index for driver performance analysis

```python
drivers_col.create_index("base_zone")

query_driver = {

"base_zone":"Central"

}

explain_driver = drivers_col.find(query_driver).explain()

print(explain_driver)
```

# Index for app performance queries

```python
app_events_col.create_index("event_type")

query_event = {

"event_type":"track_order"

}

explain_event = app_events_col.find(query_event).explain()

print(explain_event)
```

# Compare execution statistics

```python
print("Before index")

print(explain_before["executionStats"])

print("\nAfter index")

print(explain_after["executionStats"])
```
